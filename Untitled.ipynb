```python
import pandas as pd
dress = pd.read_csv("labels.csv", sep=' ')
```

```python
dress.head()
```

```python
to_drop = ['name']
dress.drop(to_drop, inplace=True, axis=1)
```

```python
dress.describe()
```

```python
dress['categories'] = dress.sum(axis=1)
```

```python
dress.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
dress.hist(figsize=(20,20), color='blue')
plt.show()
```
